# Part 3. Dictionaries and Sets

####  Use the marco geoestadístico 2010 (https://www.inegi.org.mx/app/biblioteca/ficha.html?upc=889463807469) and 2020 (https://www.inegi.org.mx/app/biblioteca/ficha.html?upc=702825292812) to obtain the “Áreas Geoestadísticas Básicas” (AGEBs) from Mérida, Yucatán. AGEBs evolve on time. Use sets to find:
    a. AGEBs which remains from 2010 on 2020.
    b. New AGEBs on 2020.
    c. AGEBs that disappear from 2010 to 2020.

```python
import geopandas as gpd

#Change paths accordignly 

file_mex = '/home/karla/Documents/UPY/HPC/CA3.-High-Performance-Python/Dictionaries_Sets/mex_2010/AGEB_urb_2010_5.shp'
file_yuc = '/home/karla/Documents/UPY/HPC/CA3.-High-Performance-Python/Dictionaries_Sets/yuc_2020/31a.shp'
```

```python
mex_2010 = gpd.read_file(file_mex)
mex_2010
```

```python
mex_2010["CVE_ENT"] = mex_2010['CVEGEO'].str.slice(0,2)
mex_2010
```

```python
mex_2010["CVE_MUN"] = mex_2010["CVEGEO"].str.slice(2,5)
mex_2010
```

```python
yuc_2010_data = mex_2010[mex_2010['CVE_ENT'] == '31']
yuc_2010_data
```

```python
mid_2010 = yuc_2010_data[yuc_2010_data["CVE_MUN"] == "050"]
mid_2010
```

```python
yuc_2020 = gpd.read_file(file_yuc)
yuc_2020
```

```python
mid_2020 = yuc_2020[yuc_2020["CVE_MUN"] == "050"]
mid_2020
```

```python
mid_2010_set = set(mid_2010['CVEGEO'])
mid_2020_set = set(mid_2020['CVEGEO'])


print("mid_2010_set:", mid_2010_set)
print("mid_2020_set:", mid_2020_set)
```

```python
remains_2010_2020 = mid_2010_set & mid_2010_set
new_2020 = mid_2020_set - mid_2010_set
disappears_2010_2020 = mid_2010_set - mid_2020_set
```

```python
remains_2010_2020_gdf = mid_2020[mid_2020["CVEGEO"].isin(remains_2010_2020)]
remains_2010_2020_gdf
```

```python
new_2020_gdf = mid_2020[mid_2020["CVEGEO"].isin(new_2020)]
new_2020_gdf
```

```python
disappears_2010_2020_gdf = mid_2010[mid_2010["CVEGEO"].isin(disappears_2010_2020)]
disappears_2010_2020_gdf
```

### Make a map with the following visualization to observe all three cases.

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

fig, ax = plt.subplots(1, 1, figsize=(15, 10))

# Plot your data with updated dark but distinct colors
remains_2010_2020_gdf.plot(ax=ax, color='black', edgecolor='white', linewidth=0.5, alpha=0.8, label='Remains 2010-2020')  # Remains dark
new_2020_gdf.plot(ax=ax, color='darkviolet', edgecolor='white', linewidth=0.5, alpha=0.8, label='New in 2020')  # Now dark purple
disappears_2010_2020_gdf.plot(ax=ax, color='darkred', edgecolor='white', linewidth=0.5, alpha=0.8, label='Disappeared 2010-2020')  # Now deep burgundy

# Create custom legend with the new colors
legend_labels = [
    Patch(facecolor='black', edgecolor='white', label='Remains 2010-2020'),
    Patch(facecolor='darkviolet', edgecolor='white', label='New in 2020'),
    Patch(facecolor='darkred', edgecolor='white', label='Disappeared 2010-2020')
]
ax.legend(handles=legend_labels)

# Customizations
ax.set_title('AGEB Changes from 2010 to 2020')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.grid(True)

plt.show()

```
